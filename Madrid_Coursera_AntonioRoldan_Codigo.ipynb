<h3> 5. Use the Foursquare API to explore the neighborhoods

```python
CLIENT_ID = '  ' # your FoursquareID
CLIENT_SECRET = '  ' # your Foursquare Secret
VERSION = '  '
LIMIT = 30
print('Your credentails:')
print('CLIENT_ID: ' + CLIENT_ID)
print('CLIENT_SECRET:' + CLIENT_SECRET)
```

<h4> Now, let's get the top 100 Movie Theaters that are within a radius of 2000 meters.

```python
LIMIT = 100
radius = 2000
categoryId= "4bf58dd8d48988d17f941735" #This is the Foursquare Category Id for Movie Theaters
venues = []

for lat, long, neighborhood in zip(Madrid_df2['Latitude'], Madrid_df2['Longitude'], Madrid_df2['Name']):
    
    # create the API request URL
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT,
        categoryId)
    
    # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    # return only relevant information for each nearby venue
    for venue in results:
        venues.append((
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
# convert the venues list into a new DataFrame
venues_df = pd.DataFrame(venues)

# define the column names
venues_df.columns = ['Neighborhood', 'Latitude', 'Longitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']

print(venues_df.shape)
venues_df
```

```python
venues_df.groupby(["Neighborhood"]).count()
```

<h3> 6. Analyze Each Neighborhood

```python
# one hot encoding
Madrid_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Madrid_onehot['Neighborhoods'] = venues_df['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Madrid_onehot.columns[-1]] + list(Madrid_onehot.columns[:-1])
Madrid_onehot = Madrid_onehot[fixed_columns]

print(Madrid_onehot.shape)
Madrid_onehot.head()
```

<h5> Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
Madrid_grouped = Madrid_onehot.groupby(["Neighborhoods"]).mean().reset_index()

print(Madrid_grouped.shape)
Madrid_grouped
```

<h4> Create a new DataFrame for Movie Theaters data only

```python
Madrid_theaters = Madrid_grouped[["Neighborhoods","Movie Theater"]]
Madrid_theaters
```

<h4> 7. Cluster Neighborhoods

<h5> Run k-means to cluster the neighborhoods in Madrid into 4 clusters.

```python
# set number of clusters
kclusters = 4

mad_clustering = Madrid_theaters.drop(["Neighborhoods"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(mad_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# create a new dataframe that includes the cluster 
mad_merged = Madrid_theaters.copy()

# add clustering labels
mad_merged["Cluster Labels"] = kmeans.labels_
```

```python
mad_merged.rename(columns={"Neighborhoods": "Name"}, inplace=True)
mad_merged.head()
```

```python
mad_merged = mad_merged.join(Madrid_df2.set_index("Name"), on="Name")
print(mad_merged.shape)
```

```python
mad_merged.head()
```

```python
# sort the results by Cluster Labels
print(mad_merged.shape)
mad_merged.sort_values(["Cluster Labels"], inplace=True)
mad_merged
```

<h5> Finally, let's visualize the resulting clusters

```python
madmap_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(mad_merged['Latitude'], mad_merged['Longitude'], mad_merged['Name'], mad_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.8).add_to(madmap_clusters)
       
madmap_clusters
```
