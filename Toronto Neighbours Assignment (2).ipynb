```python
import pandas as pd
import numpy as np
from pandas.io.html import read_html

page = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

Canada_table = pd.read_html(page, attrs={'class':'wikitable sortable'})
```

```python
Canada_df = pd.DataFrame(Canada_table[0])
Canada_df.head()
```

```python
headers = ['Postal Code', 'District', 'Neighbourhood']
```

Deleting cells with "District" = "Not assigned"

```python
indexNames = Canada_df[Canada_df['District'] == "Not assigned"].index
```

```python
Canada_df.drop(indexNames, inplace=True)
```

```python
# Grouping the same postal code Neighbours
Canada_df2 = Canada_df.groupby(['Postal Code','District'], sort=False).agg(', '.join)
Canada_df2.reset_index(inplace=True)

# Replacing the name "Not assigned" neighbourhoods with names of District
Canada_df2['Neighbourhood'] = np.where(Canada_df2['Neighbourhood'] == 'Not assigned',Canada_df2['District'], Canada_df2['Neighbourhood'])

Canada_df2
```

```python
Canada_df2.shape
```

FINAL PART 1 OF ASSIGNMENT................................................................

Import Lattitude and Longitude

```python
coord_Tor = pd.read_csv('https://cocl.us/Geospatial_data')
coord_Tor.head()
```

```python
Canada_df3 = pd.merge(Canada_df2,coord_Tor,on='Postal Code')
Canada_df3.head()
```

FINAL PART 2 OF ASSIGNMENT.................................................

CLUSTERING AND SEGMENTING NEIGHBOURHOODS

```python
Canada_final = Canada_df3[Canada_df3['District'].str.contains('Toronto',regex=False).reset_index(drop=True)]
```

```python
Canada_final2= Canada_final.reset_index(drop=True)
Canada_final2.head()
```

```python
Tor_map = folium.Map(location=[43.654260,-79.360636],zoom_start=12)

for lat,lng,district,neighbourhood in zip(Canada_final2['Latitude'],Canada_final2['Longitude'],Canada_final2['District'],Canada_final2['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, district)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='red',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.8,
    parse_html=False).add_to(Tor_map)
Tor_map
```

```python
klusters=5
toronto_clustering = Canada_final2.drop(['Postal Code','District','Neighbourhood'],1)
kmeans = KMeans(n_clusters = klusters,random_state=0).fit(toronto_clustering)
kmeans.labels_
Canada_final2.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
Canada_final2
```

```python
# create map
Tor_map_c = folium.Map(location=[43.654260,-79.360636],zoom_start=12)

# set color scheme for the clusters
x = np.arange(klusters)
ys = [i + x + (i*x)**2 for i in range(klusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(Canada_final2['Latitude'], Canada_final2['Longitude'], Canada_final2['Neighbourhood'], Canada_final2['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(Tor_map_c)
       
Tor_map_c
```
